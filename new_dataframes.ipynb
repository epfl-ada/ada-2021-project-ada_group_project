## Creation of the quote dataframes for each year:

Let's create a dataframe with about 22000 politicians with information like gender, ethnicity, party, academic degree, etc. We use the label column from this dataframe with the keywords of 'words' list to get dataframes for each year from the quotes files. 
**We show the code here however we use files that are locally stored in our computers. Indeed they were too heavy for Github. Once the dataframes are created, we add them on google drive and store them directly on our computers and github.** 
This way we avoid the processing of big data which causes issues on some computers and we can directly work with proper dataframes.

```python
import pandas as pd
```

```python
parquetdf = pd.read_parquet('../Downloads/speaker_attributes.parquet') #local path

# Extract all rows with American politicians
parquetdf2 = parquetdf[~parquetdf.occupation.isnull() & ~parquetdf.nationality.isnull()]
try1 = parquetdf2.apply(lambda x: 'Q30' in x['nationality'], axis=1)  # 'Q30' corresponds to American nationality
try2 = parquetdf2.apply(lambda x: 'Q82955' in x['occupation'], axis=1)  # 'Q82955' corresponds to politicians
try3 = np.logical_and(try1, try2)

american_politicians = parquetdf2[try3].reset_index(drop=True)

# Get rid of politicians who were born before 1935
americansnonnull = american_politicians[~american_politicians.date_of_birth.isnull()]
year = americansnonnull.apply(lambda x: x['date_of_birth'][0][1:5], axis=1)

year = year.astype(int)
todelete = year[year < 1935].index
american_politicians = american_politicians.drop(todelete)

american_politicians = american_politicians.reset_index(drop=True)
american_politicians.shape

american_politicians.head()
```

```python
names = american_politicians.label
```

```python
words = ['lesbian', 'gay', 'homosexual', 'gender', 'bisexual', 'sexuality', 'same sex', 'ally',
         'asexual', 'bi', 'biphobia', 'bisexual', 'coming out', 'coming-out', 'gender identity',
        'queer', 'genderqueer', 'gender-queer', 'homophobia', 'LGBTQ', 'LGBT', 'LGBTQ+', 'LGBTQIA',
        'lgbtq', 'lgbt', 'lgbtq+', 'lgbtqia', 'non binary', 'non-binary', 'transgender'] 
```

```python
chunk_num = 1
def process_chunk(chunk):
        print(f'Processing chunk with {len(chunk)} rows')
        
        return chunk.loc[chunk.speaker.isin(names) & (chunk.quotation.str.contains('{}'.format("|".join(words))))]
            
        

with pd.read_json('../Downloads/quotes-2018.json.bz2', lines=True, compression='bz2', chunksize=100000) as df_reader:
    for chunk in df_reader:
        processed_chunk = process_chunk(chunk)
        if (chunk_num == 1):
            dataframe18 = processed_chunk
            chunk_num += 1
        else:
            dataframe18 = pd.concat([dataframe18, processed_chunk])
        
```

```python
dataframe18.to_csv("data/df18.csv")
```
